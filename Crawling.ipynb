```python
import json
from bs4 import BeautifulSoup
import requests
import time
import re

urls_to_crawl = ["https://islamqa.info/id/categories/topics/48/perceraian-thalaq"
              "https://islamqa.info/id/categories/topics/48/perceraian-thalaq?page=2",
              "https://islamqa.info/id/categories/topics/48/perceraian-thalaq?page=3",
              "https://islamqa.info/id/categories/topics/48/perceraian-thalaq?page=4",
              "https://islamqa.info/id/categories/topics/48/perceraian-thalaq?page=5",
              "https://islamqa.info/id/categories/topics/48/perceraian-thalaq?page=6",
              "https://islamqa.info/id/categories/topics/41/mahar",
              "https://islamqa.info/id/categories/topics/41/mahar?page=2",
              "https://islamqa.info/id/categories/topics/263/sejaran-nabi",
              "https://islamqa.info/id/categories/topics/263/sejaran-nabi?page=2",
              "https://islamqa.info/id/categories/topics/263/sejaran-nabi?page=3",
              "https://islamqa.info/id/categories/topics/29/syarah-syarah-hadis",
              "https://islamqa.info/id/categories/topics/29/syarah-syarah-hadis?page=2",
              "https://islamqa.info/id/categories/topics/29/syarah-syarah-hadis?page=3",
              "https://islamqa.info/id/categories/topics/249/dakwah-kepada-orang-islam",
              "https://islamqa.info/id/categories/topics/249/dakwah-kepada-orang-islam?page=2",
              "https://islamqa.info/id/categories/topics/262/biografi",
              "https://islamqa.info/id/categories/topics/262/biografi?page=2",
              "https://islamqa.info/id/categories/topics/262/biografi?page=3"
            ]

data = []

for url in urls_to_crawl:
    # Request halaman web
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    produk_divs = soup.find_all('a', class_="post-card")

    for div in produk_divs:

        link_detail = div['href']

        judul = div.find('p', class_="has-font-content card-title").text.strip()
        time.sleep(1)
        response_detail = requests.get(link_detail)
        soup_detail = BeautifulSoup(response_detail.content, 'html.parser')
        deskripsi = soup_detail.find('div', class_="container is-large single_fatwa")
        number = deskripsi.find('div', class_="level-item has-separator")
        quest = deskripsi.find('section', class_="single_fatwa__question text-justified")
        ans = deskripsi.find('section', class_="single_fatwa__answer")
        # reference = deskripsi.find('p', class_="subtitle is-6 has-text-weight-bold is-capitalized")

        if number and quest and ans:
            nomor = number.find('p', class_="subtitle has-text-weight-bold has-title-case cursor-pointer tooltip").text.strip()
            pertanyaan = quest.find('div').text.strip()
            jawaban = ans.find('div', class_="content").text.strip()
            # referensi = reference.find('span').text.strip()

            surah_references = re.findall(r'\((QS\..+?)\)', jawaban)
            referensi = ', '.join(surah_references) if surah_references else "-"

            qa_pair = {
                "nomor": nomor,
                "link": link_detail,
                "judul": judul,
                "pertanyaan": pertanyaan,
                "jawaban": jawaban,
                "referensi": referensi
            }


            data.append(qa_pair)

json_data = json.dumps(data, indent=4, ensure_ascii=False)

print(json_data)
```

```python
import pandas as pd

df = pd.DataFrame(data)
df
```
